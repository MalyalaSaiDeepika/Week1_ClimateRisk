```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Visualize fatalities per district
plt.figure(figsize=(10,6))
sns.barplot(x='district', y='fatalities', data=df.sort_values('fatalities', ascending=False))
plt.xticks(rotation=45)
plt.title("Fatalities per District")
plt.show()
```

```python
plt.figure(figsize=(12,6))
ax = sns.barplot(
    x='district', 
    y='fatalities', 
    hue='district',  # assign x variable to hue
    data=df.sort_values('fatalities', ascending=False),
    dodge=False,      # avoids splitting bars
    palette="Reds_r"
)
plt.legend([],[], frameon=False)  # removes legend if not needed
plt.xticks(rotation=45, ha='right')
plt.title("Fatalities per District")

for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width()/2., p.get_height()),
                ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
